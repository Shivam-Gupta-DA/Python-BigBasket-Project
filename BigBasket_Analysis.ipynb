### Treating missing values

1. Product - Null values should be dropped.
2. Brand - Should be dropped
3. Rating - impute using mean/median
4. SalesPrice - impute using mean/median
5. description - should be dropped

```python
#Dropping missing values from product column 
df.dropna(subset='product',inplace=True)
df.dropna(subset='brand',inplace=True)
```

```python
df['sale_price'].fillna(df['sale_price'].median(),inplace=True)
df['rating'].fillna(df['rating'].median(),inplace=True)
```

```python
df.drop(columns=['description'],inplace=True)
```

```python
df.isnull().sum().sum()  #Here we can say the all null values are cleared
```

## Step 6. Outlier Treatment

```python
sns.pairplot(df) #Let's plot a pairplot to know how the data is plotted. 
```

#### Let's create datasets differently for the numerical and categorical columns to treat the outliers properly

```python
numer_col = [x for x in df.columns if df[x].dtypes == 'int64' or df[x].dtypes == 'float64']
numer_col
#This dataset will contain all the numerical columns
```

```python
cat_col = [x for x in df.columns if df[x].dtypes == 'O']
cat_col

#This dataset will contain all the categorical columns
```

```python
for i in numer_col:  #Let's plot the box plot for numerical columns to see the outliers
    plt.figure(figsize=(10,6))
    sns.boxplot(df[i],color='blue')
    plt.title(f'boxplot of {i}')
    plt.show()
```

```python
df1=df.copy()
```

```python
def replace_out(df,col):   #Creating a function to treat the outliers of the dataset on the basis of the median of the respective column
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    

    iqr = q3-q1

    lower_bound = q1-1.5*iqr
    upper_bound = q3+1.5*iqr

    median=df[col].median()
    df[col] = df[col].apply(lambda x: median if x < lower_bound or x > upper_bound else x )
```

```python
outliers = ['market_price','sale_price']

for i in outliers:
    replace_out(df,i)
```

```python
for i in outliers:
    plt.figure(figsize=(10,6))
    sns.boxplot(df[i])
    plt.show()
```
